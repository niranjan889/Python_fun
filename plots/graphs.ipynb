```python
%matplotlib notebook   
# %matplotlib inline
import matplotlib.pyplot as plt
from matplotlib import rc
import numpy as np
```

In the above cell, we use command '%matplotlib notebook' to make the graphs interactive. You can run all the cells and see the output below. 
If interactive graph is not desirable, then simply use '%matplotlib inline' instead.

```python
data_x= ['p@2','p@5','p@10','p@20'] # 'names for each group of plots.'
data_y1= [0.5,0.6,0.75]  
data_y2= [0.4,0.66,0.66]
data_y3= [0.23,0.5,0.61]
data_y4= [0.33,0.4,0.58]
x_labels= ['Data1','Data2','Data3','Data4']
```

```python
def bar_plot(data_x,data_y1,data_y2,data_y3,data_y4,x_labels):
    font = {'family' : 'Serif',
        'weight' : 'normal',
        'size'   : 16}
    rc('font', **font)
    
    c1='#000000'    #set color of the bar in the plot
    
    #To plot
    fig = plt.figure(num=1,figsize=(14,8))
    ax = fig.add_subplot(221)
    list_vals=np.array(data_y1)
            
    N = len(list_vals)      
    ind = np.arange(N)  # the x locations for the groups
    width = 0.4       # the width of the bars
    b=ax.bar(ind, list_vals, width, color=c1)
    
    # add some text for labels, title and axes ticks
    ax.set_ylabel('Y axis label',fontsize=18)
    ax.set_xlabel('a) '+x_labels[0],fontsize=18)
    ax.set_xticks(ind+width/2)
    ax.set_xticklabels(data_x)

    #2nd plot
    ay = fig.add_subplot(222)
    list_vals=np.array(data_y2)
    
    N = len(list_vals)      
    ind = np.arange(N)  # the x locations for the groups
    width = 0.4       # the width of the bars
    b=ay.bar(ind, list_vals, width, color=c1)
    
#     ay.set_ylim(0,0.6)
    # add some text for labels, title and axes ticks
    ay.set_ylabel('Y axis label',fontsize=18)
    ay.set_xlabel('b) '+x_labels[1],fontsize=18)
    ay.set_xticks(ind+width/2)
    ay.set_xticklabels(data_x)
    
    #3rd plot
    ay = fig.add_subplot(223)
    list_vals=np.array(data_y3)
    
    N = len(list_vals)      
    ind = np.arange(N)  # the x locations for the groups
    width = 0.4       # the width of the bars
    b=ay.bar(ind, list_vals, width, color=c1)
    
    # add some text for labels, title and axes ticks
    ay.set_ylabel('Y axis label',fontsize=18)
    ay.set_xlabel('c) '+x_labels[2],fontsize=18)
    ay.set_xticks(ind+width/2)
    ay.set_xticklabels(data_x)
    
    #4th plot
    ay = fig.add_subplot(224)
    list_vals=np.array(data_y4)
    
    N = len(list_vals)      
    ind = np.arange(N)  # the x locations for the groups
    width = 0.4       # the width of the bars
    b=ay.bar(ind, list_vals, width, color=c1)
    
    # add some text for labels, title and axes ticks
    ay.set_ylabel('Y axis label',fontsize=18)
    ay.set_xlabel('d) '+x_labels[3],fontsize=18)
    ay.set_xticks(ind+width/2)
    ay.set_xticklabels(data_x)
    
    #Adjust the spacings between the sub graphs
    plt.subplots_adjust(bottom = 0.1,top = 0.95,wspace = 0.2,hspace = 0.3 )
    plt.show()
```

```python
bar_plot(data_x, data_y1, data_y2, data_y3, data_y4,x_labels)
```
