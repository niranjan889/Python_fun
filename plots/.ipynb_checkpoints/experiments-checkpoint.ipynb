This notebook contains more advanced plots using mpld3 package

```python
# %matplotlib notebook   
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib import rc
import numpy as np
```

```python
fig1 = plt.figure()
t1 = np.arange(0.0, 2.0, 0.1)
t2 = np.arange(0.0, 2.0, 0.01)

# note that plot returns a list of lines.  The "l1, = plot" usage
# extracts the first element of the list into l1 using tuple
# unpacking.  So l1 is a Line2D instance, not a sequence of lines
l1, = plt.plot(t2, np.exp(-t2))
l2, l3 = plt.plot(t2, np.sin(2 * np.pi * t2), '--go', t1, np.log(1 + t1), '.')
l4, = plt.plot(t2, np.exp(-t2) * np.sin(2 * np.pi * t2), 'rs-.')

plt.xlabel('time')
plt.ylabel('volts')
plt.title('Damped oscillation')

plt.show()
```

```python
import mpld3
mpld3.enable_notebook()
```

```python
from mpld3 import plugins

fig, ax = plt.subplots(subplot_kw=dict(axisbg='#EEEEEE'))
ax.grid(color='white', linestyle='solid')

N = 50
scatter = ax.scatter(np.random.normal(size=N),
                     np.random.normal(size=N),
                     c=np.random.random(size=N),
                     s = 1000 * np.random.random(size=N),
                     alpha=0.3,
                     cmap=plt.cm.jet)

ax.set_title("D3 Scatter Plot", size=20)
```

```python
from mpld3 import plugins

fig, ax = plt.subplots(subplot_kw=dict(axisbg='#EEEEEE'))
ax.grid(color='white', linestyle='solid')

N = 50
scatter = ax.scatter(np.random.normal(size=N),
                     np.random.normal(size=N),
                     c=np.random.random(size=N),
                     s = 1000 * np.random.random(size=N),
                     alpha=0.3,
                     cmap=plt.cm.jet)

ax.set_title("D3 Scatter Plot (with tooltip)", size=20)

labels = ['point {0}'.format(i + 1) for i in range(N)]
# fig.plugins = plugins.PointLabelTooltip(scatter, labels)
# plugins.connect(fig, plugins.PointLabelTooltip(scatter[0]))

tooltip = plugins.PointHTMLTooltip(scatter, labels, voffset=10, hoffset=10)
plugins.connect(fig, tooltip)
```

```python
# multiple subplots, adjusted for spacing
fig = plt.figure(figsize=(8, 6))
fig.subplots_adjust(hspace=0.3)

np.random.seed(0)

for i in range(1, 5):
    ax = fig.add_subplot(2, 2, i)
    color = np.random.random(3)
    ax.plot(np.random.random(30), lw=2, c=color)
    ax.set_title("RGB = ({0:.2f}, {1:.2f}, {2:.2f})".format(*color),
                 size=14)
    ax.grid(color='lightgray', alpha=0.7)
    
mpld3.display(fig)
```
