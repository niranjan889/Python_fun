```python
%matplotlib notebook   
# %matplotlib inline
import matplotlib.pyplot as plt
from matplotlib import rc
import numpy as np

data_x=['p@2','p@5','p@10','p@20'] # 'names for each group of plots.'
data_y1=[[0.15,0.26,0.25],[0.1,0.2,0.35],[0.2,0.3,0.45],[0.3,0.52,0.55]]  
data_y2=[[0.15,0.26,0.25],[0.1,0.2,0.35],[0.2,0.3,0.45],[0.3,0.6,0.75]]
data_y3=[[0.15,0.26,0.35],[0.1,0.2,0.35],[0.2,0.3,0.45],[0.3,0.6,0.75]]
data_y4=[[0.15,0.26,0.15],[0.1,0.2,0.35],[0.2,0.3,0.45],[0.3,0.52,0.55]]

n_col = 3 #No. of column in legend
loc = 2 #Location for legend
```

```python
#Takes input as 4 lists, the output is a 2 x 2 grid.
def bar_plot2(data_x,data_y1,data_y2,data_y3,data_y4):
    font = {'family' : 'Serif',
        'weight' : 'normal',
        'size'   : 16}
    rc('font', **font)
    
    #set color of the bar in the plot
    c1='#FFFFFF'
    c2='#D1D0CE'
    c3='#777777'
    c4='#000000'
    
    #To plot
    fig = plt.figure(num=1,figsize=(14,8))
    ax = fig.add_subplot(221)
    list_vals0=np.array(data_y1[0])
    list_vals1=np.array(data_y1[1])
    list_vals2=np.array(data_y1[2])
    list_vals3=np.array(data_y1[3])
            
    N = len(list_vals0)      
    ind = np.arange(N)  # the x locations for the groups
    width = 0.2       # the width of the bars
    
    b = ax.bar(ind, list_vals0, width, color=c1)
    b1 = ax.bar(ind+0.2, list_vals1, width, color=c2)
    b2 = ax.bar(ind+0.4, list_vals2, width, color=c3)
    b3 = ax.bar(ind+0.6, list_vals3, width, color=c4)
    
    # add some text for labels, title and axes ticks
    ax.set_ylabel('Y axis label',fontsize=18)
    ax.set_xlabel('a). Model 1',fontsize=18)
    ax.set_xticks(ind+width*2)
    ax.set_xticklabels(data_x)
    ax.legend((b,b1,b2,b3), ('RC','MM','PBM','SSTREC'),ncol=n_col,prop={'size':14},frameon=False,loc=loc)

    #2nd plot
    ax = fig.add_subplot(222)
    list_vals0=np.array(data_y2[0])
    list_vals1=np.array(data_y2[1])
    list_vals2=np.array(data_y2[2])
    list_vals3=np.array(data_y2[3])
            
    N = len(list_vals0)      
    ind = np.arange(N)  # the x locations for the groups
    width = 0.2       # the width of the bars
    
    b = ax.bar(ind, list_vals0, width, color=c1)
    b1 = ax.bar(ind+0.2, list_vals1, width, color=c2)
    b2 = ax.bar(ind+0.4, list_vals2, width, color=c3)
    b3 = ax.bar(ind+0.6, list_vals3, width, color=c4)
    
    # add some text for labels, title and axes ticks
    ax.set_ylabel('Y axis label',fontsize=18)
    ax.set_xlabel('b). Model 2',fontsize=18)
    ax.set_xticks(ind+width*2)
    ax.set_xticklabels(data_x)
    ax.legend((b,b1,b2,b3), ('RC','MM','PBM','SSTREC'),ncol=n_col,prop={'size':14},frameon=False,loc=loc)
    
    #3rd plot
    ax = fig.add_subplot(223)
    list_vals0=np.array(data_y3[0])
    list_vals1=np.array(data_y3[1])
    list_vals2=np.array(data_y3[2])
    list_vals3=np.array(data_y3[3])
            
    N = len(list_vals0)      
    ind = np.arange(N)  # the x locations for the groups
    width = 0.2       # the width of the bars
    
    b = ax.bar(ind, list_vals0, width, color=c1)
    b1 = ax.bar(ind+0.2, list_vals1, width, color=c2)
    b2 = ax.bar(ind+0.4, list_vals2, width, color=c3)
    b3 = ax.bar(ind+0.6, list_vals3, width, color=c4)
    
    # add some text for labels, title and axes ticks
    ax.set_ylabel('Y axis label',fontsize=18)
    ax.set_xlabel('c). Model 3',fontsize=18)
    ax.set_xticks(ind+width*2)
    ax.set_xticklabels(data_x)
    ax.legend((b,b1,b2,b3), ('RC','MM','PBM','SSTREC'),ncol=n_col,prop={'size':14},frameon=False,loc=loc)
    
    #4th plot
    ax = fig.add_subplot(224)
    list_vals0=np.array(data_y4[0])
    list_vals1=np.array(data_y4[1])
    list_vals2=np.array(data_y4[2])
    list_vals3=np.array(data_y4[3])
            
    N = len(list_vals0)      
    ind = np.arange(N)  # the x locations for the groups
    width = 0.2       # the width of the bars
    
    b = ax.bar(ind, list_vals0, width, color=c1)
    b1 = ax.bar(ind+0.2, list_vals1, width, color=c2)
    b2 = ax.bar(ind+0.4, list_vals2, width, color=c3)
    b3 = ax.bar(ind+0.6, list_vals3, width, color=c4)
    
    # add some text for labels, title and axes ticks
    ax.set_ylabel('Y axis label',fontsize=18)
    ax.set_xlabel('d). Model 4',fontsize=18)
    ax.set_xticks(ind+width*2)
    ax.set_xticklabels(data_x)
    ax.legend((b,b1,b2,b3), ('RC','MM','PBM','SSTREC'),ncol=n_col,prop={'size':14},frameon=False,loc=loc)
    
    plt.subplots_adjust(bottom = 0.1,top = 0.95,wspace = 0.2,hspace = 0.3 )
    plt.show()

bar_plot2(data_x, data_y1, data_y2, data_y3, data_y4)
```
